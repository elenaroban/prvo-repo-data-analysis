```python
import pandas as pd
```

```python
df = pd.read_csv('data_cleaning_dataset.csv')
df.head()
```

```python
df.isna().sum()
```

```python
df['Quantity'].fillna(1, inplace=True)
df['Category'].fillna('Unknown', inplace=True)
df['Supplier'].fillna('No name', inplace=True)
df['Price'].fillna(100, inplace=True)
df['Discount'].fillna(10, inplace=True)
df.isna().sum()
```

```python
df
```

```python
df['Quantity'] = df['Quantity'].astype(int)
df.info()
```
