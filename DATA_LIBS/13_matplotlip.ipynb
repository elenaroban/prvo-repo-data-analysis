```python
df = pd.read_csv('datasets/sales.csv')
df.head()
```

```python
df.plot(x='Производ', y='Продажби', kind = 'bar')
```

```python
df_grouped = df.groupby('Производ')
```

```python
data_sum =df_grouped['Продажби'].sum()
```

```python
data_sum.plot(kind='bar')
plt.xlabel('Производ')
plt.ylabel('Број на продажби')
plt.savefig('prodazbi.png')
```

```python
products =list(data_sum.index) #lista od grupirani produkti
sum_of_sales = data_sum.to_list() #lista od zbirovi na prodazbi na produkti
```

```python
plt.barh(y=products, width=sum_of_sales, color = '#cc6699')

plt.xlabel('Proizvodi')
plt.ylabel('Zbir na prodazbi')
plt.title('Prodazba na razni proizvodi')
plt.xticks(rotation = 90, fontsize=13, color='red')

plt.show()
```

```python
fig, axis= plt.subplots(2,2, figsize=(15,10))
```

```python
axis[0,0].bar(x=products, height=sum_of_sales, color='#cc6699')
axis[0,1].barh(y=products, width=sum_of_sales, color='blue')
axis[1,0].pie(sum_of_sales, labels=products)
fig.savefig('complex_chart.png')

fig
```
